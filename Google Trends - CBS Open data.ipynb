```python
from pytrends.request import TrendReq

pytrends = TrendReq()
```

```python
kw_list = ["اللجوء"]
pytrends.build_payload(kw_list, cat=0, timeframe='all', geo='SY', gprop='')
trends = pytrends.interest_over_time().اللجوء
print(trends)
```

```python
import pandas as pd
trends.plot()
```
